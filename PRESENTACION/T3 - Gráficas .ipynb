```python
import pandas as pd
import matplotlib.pyplot as plt #Importar libreria para graficar resultados
```

```python
mainpath = "C:/Users/iRed-Saul/Documents/GitHub/python-ml-course/datasets" #lectura absoluta de la direccion
#renombre de variables para direccion de archivos
filename = "customer-churn-model/Customer Churn Model.txt"
fullpath = mainpath + "/" + filename
df = pd.read_csv(fullpath)
df.head(100) #impresion de variables
```

### Scatter Plot // Nube de dispersion

```python
df.plot(kind="scatter",x="Night Mins",y="Night Charge") #realcion de uso y carga de noche
```
